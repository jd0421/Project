### · 월 별 Conversion Rate

````python
# 월 별 Conversion Rate (1)
# 거래 Revenue를 count로 하는게 맞는지 확인, 전환 수 기준인 것으로 보아 count 가 맞는 것 같습니다.
df_cov = pd.DataFrame(df.groupby(df['date'].dt.strftime('%Y-%m'))[['sessionId', 'totals_transactionRevenue']].agg(['count'])).reset_index()
df_cov.columns = ['date', 'total', 'Revenue']

# 월 별 Conversion Rate (2)
df_cov['cov_rate'] = round((df_cov['Revenue']/df_cov['total'])*100,2)
df_cov
````

````python
# 월 별 Conversion Rate 시각화
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')

plt.figure(figsize=(12,5))
bars = plt.bar(df_cov.date, df_cov['cov_rate'])

# 데이터 레이블 추가
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height,
             f'{float(height)}',
             ha='center', va='bottom', fontsize=9)

plt.xticks(rotation=0)
plt.xlabel('Date')
plt.ylabel('Value')
plt.tight_layout()
plt.show()

# 전환율 kpi는 최상단 3개 성과 기준 설정 제안
#
````

````python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# device, browser, OS 카테고리 기준 transactionRevenue 추이 변화 (중앙값 기준)

# 첫 번째 그래프 데이터
df_device_browser = pd.DataFrame(df.groupby(df['device_browser'])['totals_transactionRevenue'].mean()).reset_index()
df_device_browser = df_device_browser.dropna().sort_values("totals_transactionRevenue", ascending = True)

# 두 번째 그래프 데이터
df_device_category = pd.DataFrame(df.groupby(df['device_deviceCategory'])['totals_transactionRevenue'].mean()).reset_index()
df_device_category = df_device_category.dropna().sort_values("totals_transactionRevenue", ascending = True)

# 세 번째 그래프 데이터
df_device_os = pd.DataFrame(df.groupby(df['device_operatingSystem'])['totals_transactionRevenue'].mean()).reset_index()
df_device_os = df_device_os.dropna().sort_values("totals_transactionRevenue", ascending = True)

# 1x2 서브플롯 생성
fig = make_subplots(rows=1, cols=3, subplot_titles=('Transaction Revenue by Device Category', 'Transaction Revenue by Device Browser', 'Transaction Revenue by OS'))

# `Device Category` 그래프를 첫 번째 서브플롯에 추가 (수평 막대그래프로 변경)
fig.add_trace(
    go.Bar(x=df_device_category['totals_transactionRevenue'], y=df_device_category['device_deviceCategory'], orientation='h'),
    row=1, col=1
)

# `Device Browser` 그래프를 두 번째 서브플롯에 추가 (수평 막대그래프로 변경)
fig.add_trace(
    go.Bar(x=df_device_browser['totals_transactionRevenue'], y=df_device_browser['device_browser'], orientation='h'),
    row=1, col=2
)

# 'Operating System' > 세 번째 서브플롯에 추가
fig.add_trace(
    go.Bar(x=df_device_os['totals_transactionRevenue'], y=df_device_os['device_operatingSystem'], orientation='h'),
    row=1, col=3
)

# 레이아웃 업데이트 및 출력
fig.update_layout(title_text="Transaction Revenue Analysis")
plt.tight_layout()
fig.show()

## 기기 기준 데스크톱에서 고매출을 보였으며, 모바일, 태블릿의 2배 이상 수준, web친화적인 매출 전략 수립을 권장드리나, MO 환경은 상거래 시장에서 높은 시장성을 가진 기기 환경임을 감안하여, 점차 매출 비중을 높일 수 있는 방향에 대한 추가 고민이 필요할 것으로 사료
## 브라우저 기준 대중적인 브라우저에서 높은 거래량 파악, 다만 Firefox 사용률이 높은 이유와, 타 브라우저의 추가적 최적화 필요 여부에 대해서 검토 필요
## chorme OS 에서 높은 거래가 발생되었는데, 이에 대한 추가 원인 파악 필요 사료됩니다. (windows, mac OS 외 대중적인 PC, MO OS 외에 높은 비중을 차지한 원인 파악)
````

````python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots


# 첫 번째 그래프 데이터
df_trafficSource_campaign = pd.DataFrame(df.groupby(df['trafficSource_campaign'])['totals_transactionRevenue'].mean()).reset_index()
df_trafficSource_campaign = df_trafficSource_campaign.dropna().sort_values("totals_transactionRevenue", ascending = True)

# 두 번째 그래프 데이터
df_device_category = pd.DataFrame(df.groupby(df['trafficSource_medium'])['totals_transactionRevenue'].mean()).reset_index()
df_device_category = df_device_category.dropna().sort_values("totals_transactionRevenue", ascending = True)

# 3
df_trafficSource_source = pd.DataFrame(df.groupby(df['trafficSource_source'])['totals_transactionRevenue'].mean()).reset_index()
df_trafficSource_source = df_trafficSource_source.dropna().sort_values("totals_transactionRevenue", ascending = True)[:10]

# 1x3 서브플롯 생성
fig = make_subplots(rows=1, cols=3, subplot_titles=('trafficSource_medium', 'trafficSource_campaign', 'trafficSource_source'))

# `Device Category` 그래프를 첫 번째 서브플롯에 추가 (수평 막대그래프로 변경)
fig.add_trace(
    go.Bar(x=df_device_category['totals_transactionRevenue'], y=df_device_category['trafficSource_medium'], orientation='h'),
    row=1, col=1
)

# `Device Browser` 그래프를 두 번째 서브플롯에 추가 (수평 막대그래프로 변경)
fig.add_trace(
    go.Bar(x=df_trafficSource_campaign['totals_transactionRevenue'], y=df_trafficSource_campaign['trafficSource_campaign'], orientation='h'),
    row=1, col=2
)

# 3
fig.add_trace(
    go.Bar(x=df_trafficSource_source['totals_transactionRevenue'], y=df_trafficSource_source['trafficSource_source'], orientation='h'),
    row=1, col=3
)

# 레이아웃 업데이트 및 출력
fig.update_layout(title_text="Transaction Revenue Analysis")
fig.show()

## 노출 목적 광고에서 가장 많은 매출 발생, cpc 목적 대비 5배 이상, 전체 광고 예산 비중을 노출 목적 광고에 집중 제안, 추가 매출 발생 효과 기대
## AW 계열의 캠페인에서 고매출 발생, 기존 캠페인들의 Accessories, Apparel 카테고리 타겟 전환을 통해 매출 증대 기회를 도모해볼 수 있겠으며 , 동적광고(Dynamic Search Ads Whole Site) 매출 발생도 높은 바, 동적광고 세팅 확장에 대한 검토 제안
##
````
