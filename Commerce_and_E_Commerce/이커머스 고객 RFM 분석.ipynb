### · Recency(최근성)
---

````python
df.head(5)
````

````python
# split 사용해서 송장 데이터에 날짜 데이터만 추출하기
# InvoiceDate 컬럼은 현재 object type이므로, split 함수 사용, 0 번째 건만 추출
df['InvoiceDate'].str.split(' ').str[0]
````

````python
# Datatime index
# to_datetime 으로 추출
import pandas as pd
df['Date'] = pd.to_datetime(df['InvoiceDate'].str.split(' ').str[0])
df.head(5)
````

````python
df['Date'].min(), df['Date'].max()
````

````python
# 고객ID별 가장 마지막 구매일
# as_index=False 사용 통해 groupby 이후 reset_index 과정을 함께 진행할 수 있습니다.
recency_df = df.groupby('CustomerID',as_index=False)['Date'].max()
recency_df.columns = ['CustomerID','LastPurchaseDate']
recency_df.head()
````

````python
# 고객의 가장 마지막 구매일로 부터 몇일이 지났는지를 계산하기 위함
recency_df['Recency'] = recency_df['LastPurchaseDate'].apply(lambda x : (df['Date'].max() - x).days) # 전체 일자로부터 가장 최근일 (Date 컬럼의 max) 에서 고객별 마지막 구매일을 제외한 일자 수
recency_df.drop(columns=['LastPurchaseDate'],inplace=True) # 기존 마지막구매일 컬럼을 제외
````

````python
recency_df.head(5)
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use(['dark_background'])

# Recency 별 분포는 어떻게 될까
# 100일 이하의 Proportion 이 높은 비중을 차지하는 것으로 확인 되었습니다.
sns.displot(data = recency_df, x="Recency", bins = 50);
plt.gcf().set_size_inches(16.5, 3)
````

````python
# 동일한 비중으로 cut하는 방향 ? 일자 절대값을 지정하여 cut하는 방향?
# 일자 절대값으로 지정하여 cut 하는 방향이 좋을 것 같습니다. 현재의 분포를 고려하여, 각 단계를 동일한 비중 구분 시, 고객 성향에 대한 왜곡가능성 예상됩니다.
# 50일 전과 50일 이후라고 가정, 두 고객의 절대 Recency 일자 기준으로는 유사하다고 할 수 있으나, 비중 기준으로 50일을 설정할 경우 서로 다른 계급으로 판정되며, 이는 50일 이후 고객의 가능성을 왜곡하여 해석할 수 있다는 위험성이 있습니다.

# 다만 효율 고려 방향이라면 동일한 proportion을 가진 bins로 접근하는 것이 적절하다고 사료되며, 이를 통해 절대 bins로 나눈 비중 대비, 효율적인 리소스 운영이 가능할 것으로 사료됩니다.

# quantity를 중요로 한다면 절대값으로, quality를 중요로 한다면 비중으로 접근

# 아래는 총 5개의 동일 비중으로 quarter를 나눈 후 각 구분별 count 수 입니다.
pd.qcut(recency_df['Recency'], q=5).value_counts().sort_index()
````

### · Frequency(최빈성)
---

````python
# Customer ID당 유니크한 Invoice를 1개의 주문건으로 인식하여 얼마나 자주 구매하고 있는지를 파악
frequency_df = df.copy()
frequency_df.drop_duplicates(subset=['CustomerID','InvoiceNo'], keep="first", inplace=True)
frequency_df = frequency_df.groupby('CustomerID', as_index=False)['InvoiceNo'].count()
frequency_df.columns = ['CustomerID','Frequency']
frequency_df.head()
````

````python
sns.displot(data = frequency_df, x="Frequency", bins = 50);
plt.gcf().set_size_inches(16.5, 3)
````

### · Monetary(금액)
---

````python
# 구매금액 = 구매개수 * 구매단가
df['Total_cost'] = df['UnitPrice'] * df['Quantity']
monetary_df=df.groupby('CustomerID',as_index=False)['Total_cost'].sum()
monetary_df.columns = ['CustomerID','Monetary']
monetary_df.head()
````

````python
# Data merge

# recency and frequency
rf = recency_df.merge(frequency_df,how='left',on='CustomerID')

# monetary

rfm = rf.merge(monetary_df,how='left',on='CustomerID')

rfm.head(5)
````

## **🔊Process03**  
**┗ RFM 활용 서비스 이용 수준 측정**  
---



### · Scailing Data
---

````python
# Min max scale = 최대값을 1, 최소값을 0으로 표준화하는 기법
from sklearn.preprocessing import minmax_scale

# 최근성은 숫자가 작을수록, 즉 최근 구매일이 얼마 지나지 않은 고객이 더 점수가 높음
# 적은게 더 좋다
# 1 - 계산값 통해 적은 일자가 높은 값을 갖도록 역산한다
rfm['Recency'] = minmax_scale(rfm['Recency'], axis=0, copy=True)
rfm['Recency'] = 1-rfm['Recency']

rfm['Frequency'] = minmax_scale(rfm['Frequency'], axis=0, copy=True)

rfm['Monetary'] = minmax_scale(rfm['Monetary'], axis=0, copy=True)

# Score
rfm['Score']=rfm['Recency']+rfm['Frequency']+rfm['Monetary']

# Score scaling 100 socre
rfm
````

````python
# 100점을 곱해서 100점 만점으로 scaling
rfm['Score']=minmax_scale(rfm['Score'], axis=0, copy=True)*100 # r, f m scoring 값의 합산은 1이 넘을 수 있으므로, 추가적인 scaling 작업을 수행합니다.
rfm['Score']=round(rfm['Score'],0)
````

### · Grade(점수) 구간화
---

````python
sns.displot(rfm['Score']);
````

````python
rfm.head(5)
````

````python
# 점수별로 고객의 등급을 부여
# 60점 이상 Very storong
# 40~60점 Strong
# 20~40점 Normal
# 10~20점 Weak
# 0~10점 Very Weak

rfm['Grade'] = rfm['Score'].apply(lambda x : '01.Very Strong' if x>=60 else
                                  ('02.Strong' if x>=40 else
                                   ('03.Normal' if x>=20 else
                                    ('04.Weak' if x>=10 else '05.Very Weak'))))
rfm.head(5)
````

#### · 월 별 Grade 이용 비중
---

````python
df_grade = df.groupby(['Date_1', 'Grade'],as_index=False)['CustomerID'].nunique()
df_grade.head(5)
````

````python
df_pivot = pd.pivot_table(df_grade,            # 피벗할 데이터프레임
                      index = 'Date_1',        # 행 위치에 들어갈 열
                      columns = 'Grade',       # 열 위치에 들어갈 열
                      values = 'CustomerID')   # 데이터로 사용할 열

df_pivot.fillna(0, inplace=True)
df_pivot['total'] = df_pivot['01.Very Strong'] + df_pivot['02.Strong'] + df_pivot['03.Normal'] + df_pivot['04.Weak'] + df_pivot['05.Very Weak']


df_pivot.iloc[:,0] = (df_pivot.iloc[:,0] / df_pivot['total'])
df_pivot.iloc[:,1] = (df_pivot.iloc[:,1] / df_pivot['total'])
df_pivot.iloc[:,2] = (df_pivot.iloc[:,2] / df_pivot['total'])
df_pivot.iloc[:,3] = (df_pivot.iloc[:,3] / df_pivot['total'])
df_pivot.iloc[:,4] = (df_pivot.iloc[:,4] / df_pivot['total'])

df_pivot.drop(['total'], axis=1, inplace=True)
df_pivot
````

````python
import warnings
warnings.filterwarnings('ignore')
plt.style.use(['dark_background'])

ax = df_pivot.plot(kind='barh', stacked=True, title="years amt", rot=0);
for p in ax.patches:
    left, bottom, width, height = p.get_bbox().bounds
    ax.annotate("%.1f"%(width*100), xy=(left+width/2, bottom+height/2), ha='center', va='center', color='black');

plt.box(False)
plt.gcf().set_size_inches(25, 15)
plt.legend(bbox_to_anchor=(1.04,1))
plt.tight_layout()
plt.show()
````
